​​​​​​​​adjusted profitability = (revenue-cost) * (1-risk factor) + engagement bonus
or
profit=revenue-cost-risk+engagement

```python
interest=['Average Revolve Balance in last 12m','Total Consumer Lend Line Amount','Count of Active Charge Cards','Total Lend Line Amount']
moneyflow=['Total Spend in last 12m','Airlines Spend in 12m','Other Spend in 12m','Entertainment Spend in 12m','Lodging Spend in 12m','Dining Spend in 12m']
cost=['Lounge Access Count','Entertainment Credit Used Amount','Credits used in airlines','Cab benefits usage','Rewards Points Balance','Rewards point redeemed in 12months']
engagement=['Number of Supplementary Accounts','Emails Open in Last 6 months','Emails Clicked in Last 6 months','Login Counts to website']
risk=['Cancellation Calls in last 12m','Cancellation Calls due to Collection','Average Risk Score in 12m']
```

```python
print(new['Divergence'].where(new['Feature'].isin(interest)).sum(),
new['Divergence'].where(new['Feature'].isin(moneyflow)).sum(),
new['Divergence'].where(new['Feature'].isin(cost)).sum(),
new['Divergence'].where(new['Feature'].isin(engagement)).sum(),
new['Divergence'].where(new['Feature'].isin(risk)).sum())
```

```python
weights_interest=[]
for c in interest:
    weights_interest.append(new['Divergence'].where(new['Feature']==c).sum()/new['Divergence'].where(new['Feature'].isin(interest)).sum())
weights_moneyflow=[]
for c in moneyflow:
    weights_moneyflow.append(new['Divergence'].where(new['Feature']==c).sum()/new['Divergence'].where(new['Feature'].isin(moneyflow)).sum())
weights_cost=[]
for c in cost:
    weights_cost.append(new['Divergence'].where(new['Feature']==c).sum()/new['Divergence'].where(new['Feature'].isin(cost)).sum())
weights_engagement=[]
for c in engagement:
    weights_engagement.append(new['Divergence'].where(new['Feature']==c).sum()/new['Divergence'].where(new['Feature'].isin(engagement)).sum())
weights_risk=[]
for c in risk:
    weights_risk.append(new['Divergence'].where(new['Feature']==c).sum()/new['Divergence'].where(new['Feature'].isin(risk)).sum())
```

```python
print(weights_interest,weights_moneyflow,weights_cost,weights_engagement,weights_risk)
```

```python
Interest=0
MoneyFlow=0
Cost=0
Engagement=0
Risk=0
for i in range(len(interest)):
    Interest=Interest+weights_interest[i]*a[interest[i]]
for i in range(len(moneyflow)):
    MoneyFlow=MoneyFlow+weights_moneyflow[i]*a[moneyflow[i]]
for i in range(len(cost)):
    Cost=Cost+weights_cost[i]*a[cost[i]]
for i in range(len(engagement)):
    Engagement=Engagement+weights_engagement[i]*a[engagement[i]]
for i in range(len(risk)):
    Risk=Risk+weights_risk[i]*a[risk[i]]
```

```python
Profitability1=(Interest+MoneyFlow-Cost) * (1-Risk) + Engagement 
Profitability=Interest+MoneyFlow+Engagement-Risk-Cost
Profitability
```
