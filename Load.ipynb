```python
columns=['Total Spend in last 12m','Average Risk Score in 12m','Credits used in airlines','Entertainment Credit Used Amount']
for c in columns:
    df[c].fillna(df[c].median(),inplace=True)


columns2=['Login Counts to website','Lounge Access Count','Cab benefits usage','Number of Supplementary Accounts','Count of Active Charge Cards']
for c in columns2:
    df[c].fillna(df[c].mode()[0],inplace=True)
```

```python
s=['Airlines Spend in 12m','Other Spend in 12m','Entertainment Spend in 12m','Lodging Spend in 12m','Dining Spend in 12m']
from IPython.display import display
for c in s:
    for a in s:
        if c!=a:
            display(df[df[c].notnull() & df[a].isnull()][[c,a]])
```

```python
for c in s:
    df[c].fillna(0,inplace=True)
```

```python
df['Emails Open in Last 6 months'].fillna(df['Emails Open in Last 6 months'].mode()[0],inplace=True)
df['Emails Clicked in Last 6 months'].fillna(0,inplace=True)
```

```python
df[df['Total Lend Line Amount'].notnull() & df['Total Consumer Lend Line Amount'].isnull()][['Total Lend Line Amount','Total Consumer Lend Line Amount']]
```
